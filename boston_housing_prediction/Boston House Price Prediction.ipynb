# Simple Linear Regression (We want to predict MEDV, or the price)

```python
X = dataset['RM'].values.reshape(-1, 1)
```

```python
y = dataset['MEDV'].values
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(X, y)
```

```python
plt.figure(figsize = (10, 8))
plt.scatter(X, y, color = 'red')
plt.plot(X, regressor.predict(X), color = 'green')
plt.xlabel('Average number of rooms per dwelling')
plt.ylabel('Median value of owner-occupied homes in $1000')
plt.show()
```

```python
plt.figure(figsize = (10, 8))
sns.regplot(X, y)
plt.xlabel('Average number of rooms per dwelling')
plt.ylabel('Median value of owner-occupied homes in $1000')
plt.show()
```

# Robust Regression
## RANSAC Regression

```python
from sklearn.linear_model import RANSACRegressor
```

```python
regressor = RANSACRegressor()
```

```python
X = dataset['RM'].values.reshape(-1, 1)
```

```python
y = dataset['MEDV'].values
```

```python
regressor.fit(X, y)
```

```python
inlier_mask = regressor.inlier_mask_
outlier_mask = np.logical_not(inlier_mask)
```

```python
line_x = np.arange(3, 10, 1)
line_y = regressor.predict(line_x.reshape(-1, 1))
```

```python
plt.figure(figsize = (12, 10))
plt.scatter(X[inlier_mask], y[inlier_mask], color = 'blue')
plt.scatter(X[outlier_mask], y[outlier_mask], color = 'green')
plt.plot(line_x, line_y, color = 'red')
plt.xlabel('Average number of rooms per dwelling')
plt.ylabel('Median value of owner occupied rooms in 1000$')
plt.legend(loc = 'upper left')
plt.show()
```
